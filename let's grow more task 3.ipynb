```python
import cv2
```

# READ AND DISPLAY IMAGE IN RGB FORMAT


```python
image = cv2.imread("Itachi.jpg")
cv2.imshow("Itachi", image)
cv2.waitKey(0)
```

# PLOT IMAGE

```python
import matplotlib.pyplot as plt
%matplotlib inline
img_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

# READ AND DISPLAY GRAY SCALE IMAGE

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("New Tiger", gray_image)
cv2.waitKey(0)
```

# PLOT GRAY IMAGE

```python
plt.imshow(gray_image)
```

# plt.imshow(gray_image)

```python
inverted_image = 255 - gray_image
cv2.imshow("Inverted", inverted_image)
cv2.waitKey(0)
```

# PLOT THE INVERTED GRAY SCALE IMAGE

```python
plt.imshow(inverted_image)
```

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

# CONVERTING INVERTED GRAY SCALE IMAGE TO PENCIL SKETCH


```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```

# PLOTING PENCIL SKETCH

```python
plt.imshow(pencil_sketch)
```

# ORIGINAL IMAGE AND PENCIL SKETCH

```python
cv2.imshow("original image", image)
cv2.imshow("pencil sketch", pencil_sketch)
cv2.waitKey(0)
```
